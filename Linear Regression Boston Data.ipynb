Linear regression model using scikit learn in boston data to predict
'Price' based on other dependent variable.

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import stats
from sklearn.datasets import load_boston
import seaborn as sns
```

```python
#loading the dataset
boston=load_boston()
```

```python
#converting data into DataFrame
data=pd.DataFrame(boston.data)
```

```python
data.head()
```

```python
data.head(1)
```

```python
boston.feature_names
```

```python
#Adding respective data Columns to fdataframe
data.columns=boston.feature_names
data.head()
```

```python
#Adding Target Column to dataframe
data['PRICE']=boston.target
```

```python
data.head()
```

```python
#plotting the distribution plots
plt.figure(figsize=(20,25))
plotnumber = 1

for column in data:
        ax = plt.subplot(4,4,plotnumber)
        sns.distplot(x=data[column])
        plotnumber+=1
```

```python
#dependent variables
X=data.iloc[:,:-1]
#independent variable
y=data.iloc[:,-1]
```

```python
#finding null values
data.isnull().sum()
```

```python
#plotting scatter plots
plt.figure(figsize=(20,30), facecolor='white')
plotnumber = 1

for column in X:
        ax = plt.subplot(5,3,plotnumber)
        plt.scatter(X[column],y)
        plotnumber+=1
        
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#splitting dataframe into traing and test data
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=32)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#strandarising the data
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
#training the model with train data
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
#accuracy of model for train data 
model.score(X_train,y_train)
```

```python
#accuracy of model for test data
model.score(X_test,y_test)
```

# RIDGE Regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge=Ridge()
```

```python
parameters={'alpha':[1e-15,1e-10,1e-5,1e-2,1,2,5,9,10,20,30,50,60,70,80,100,200,250,300]}
```

```python
ridge_regg=GridSearchCV(ridge,parameters,scoring='neg_mean_squared_error',cv=5)
ridge_regg.fit(X,y)
```

```python
ridge_regg.best_score_
```

```python
ridge_regg.best_params_
```

```python
data.head(1)
```

```python
ridge_regg.predict([[0.00632,18.0,2.31,0.0,0.538,6.575,65.2,4.09,1.0,296.0,15.3,396.9,4.98]])[0]
```

```python
# saving the model to the local file system
import pickle
filename = 'boston_linear.pickle'
pickle.dump(ridge_regg, open(filename, 'wb'))
```

```python
loaded_model = pickle.load(open(filename, 'rb'))
```

```python
loaded_model.predict([[0.00632,18.0,2.31,0.0,0.538,6.575,65.2,4.09,1.0,296.0,15.3,396.9,4.98]])[0]
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
```

```python
lasso=Lasso()
parameters={'alpha':[1e-15,1e-10,1e-5,1e-2,1,2,5,9,10,20,30,50,60,70,80,100,200,250,300]}
lasso_regg=GridSearchCV(lasso,parameters,scoring='neg_mean_squared_error',cv=5)
lasso_regg.fit(X,y)
```

```python
lasso_regg.best_score_
```

```python
lasso_regg.best_params_
```
